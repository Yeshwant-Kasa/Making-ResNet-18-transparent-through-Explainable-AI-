```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader, Dataset, random_split
from captum.attr import Saliency, IntegratedGradients, LayerGradCam
from tensorflow.keras.datasets import cifar10
import tensorflow as tf

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# Custom Dataset (smaller size and less images for memory stability)
class CIFAR10Dataset(Dataset):
    def __init__(self, classes, train=True, limit=3000):
        (x_train, y_train), (x_test, y_test) = cifar10.load_data()
        x = x_train if train else x_test
        y = y_train if train else y_test
        y = y.flatten()
        mask = np.isin(y, classes)
        x, y = x[mask][:limit], y[mask][:limit]
        self.x = x
        self.y = [classes.index(label) for label in y]
        self.transform = lambda img: tf.image.resize(img, [128,128]).numpy() / 255.0

    def __len__(self):
        return len(self.y)

    def __getitem__(self, idx):
        img = self.transform(self.x[idx])
        img = np.transpose(img, (2,0,1)).astype(np.float32)
        return torch.tensor(img), torch.tensor(self.y[idx])

# Load reduced datasets (fewer images, smaller size)
classes = [0, 1, 2]
train_dataset = CIFAR10Dataset(classes, train=True, limit=3000)
test_dataset = CIFAR10Dataset(classes, train=False, limit=900)

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True, num_workers=2)
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False, num_workers=2)

# Load ResNet-18 (without torchvision)
model = torch.hub.load('pytorch/vision:v0.15.2', 'resnet18', pretrained=True)
model.fc = nn.Linear(model.fc.in_features, 3)
model = model.to(device)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

epochs = 5  # fewer epochs to guarantee stability first
train_loss, train_acc, val_acc = [], [], []

for epoch in range(epochs):
    model.train()
    total_loss, correct, total = 0, 0, 0
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        _, predicted = torch.max(outputs, 1)
        correct += (predicted == labels).sum().item()
        total += labels.size(0)
    train_loss.append(total_loss / len(train_loader))
    train_acc.append(correct / total)

    # Validate
    model.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, predicted = torch.max(outputs, 1)
            correct += (predicted == labels).sum().item()
            total += labels.size(0)
    val_acc.append(correct / total)

    print(f"Epoch [{epoch+1}/{epochs}] | Loss: {train_loss[-1]:.4f} | Train Acc: {train_acc[-1]:.4f} | Val Acc: {val_acc[-1]:.4f}")

# Plot results
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(train_loss, label='Train Loss')
plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend()

plt.subplot(1,2,2)
plt.plot(train_acc, label='Train Acc')
plt.plot(val_acc, label='Validation Acc')
plt.xlabel('Epoch'); plt.ylabel('Accuracy'); plt.legend()
plt.show()



```

```python
def visualize(model, input_tensor, label):
    saliency = Saliency(model).attribute(input_tensor, target=label).detach().cpu().numpy()[0]
    ig = IntegratedGradients(model).attribute(input_tensor, target=label, n_steps=20).detach().cpu().numpy()[0]
    gradcam = LayerGradCam(model, model.layer4[-1]).attribute(input_tensor, target=label).detach().cpu().numpy()[0, 0]

    input_img = input_tensor.cpu().numpy()[0].transpose(1, 2, 0)

    plt.figure(figsize=(15,4))
    plt.subplot(1,4,1)
    plt.imshow(input_img); plt.title("Original"); plt.axis('off')

    plt.subplot(1,4,2)
    plt.imshow(np.abs(saliency).max(axis=0), cmap='hot'); plt.title("Saliency"); plt.axis('off')

    plt.subplot(1,4,3)
    plt.imshow(gradcam, cmap='jet'); plt.title("Grad-CAM"); plt.axis('off')

    plt.subplot(1,4,4)
    plt.imshow(np.abs(ig).max(axis=0), cmap='hot'); plt.title("Integrated Gradients"); plt.axis('off')

    plt.show()

    # Example visualization
model.eval()
example_img, example_lbl = next(iter(test_loader))
visualize(model, example_img[0].unsqueeze(0).to(device), example_lbl[0].item())
```
